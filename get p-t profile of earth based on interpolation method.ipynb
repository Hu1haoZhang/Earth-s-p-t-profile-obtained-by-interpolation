Now we use interpolation to derive the required p-t profile in logspace

Here I need a value of temperature for a pressure distribution of logspace (-5, 0, 60)

```python
p_log = np.log10(pre)
n_p_log = np.log10(np.logspace(-5, 0, 60)) #You can adjust this section to suit your needs
```

```python
x = n_p_log
xp = np.flip(p_log)
fp = np.flip(tre) 
y = np.interp(x, xp, fp)
plt.plot(xp, fp, '-o') #Basic Data
plt.plot(x, y, 'x')   #Simulation data
plt.show()
```
